```python
'''1. Write a python program to calculate the area and perimeter of the following figures.
a. Circle
b. Triangle
c. Rectangle'''

def circle_area(r):
    pi=3.14
    area_circle=pi*r*r
    return area_circle

def triangle_area(b,h):
    area_triangle=.5*b*h
    return area_triangle

def rectangle_area(B,H):
    area_rectangle=B*H
    return area_rectangle


def circle_peri(r):
    pi=3.14
    peri_circle=2*pi*r
    return peri_circle

def triangle_peri(a,b,c):
    peri_triangle=a+b+c
    return peri_triangle

def rectangle_peri(B,H):
    peri_rectangle=2*(B+H)
    return peri_rectangle

r=float(input("Enter the radius of circle :"))
print("The area of circle is :",circle_area(r))
print("The perimeter of circle is :",circle_peri(r))

b=float(input("Enter the base of triangle :"))
h=float(input("Enter the height of triangle :"))
print("The area of circle is :",triangle_area(b,h))
a=float(input("Enter the base of triangle :"))
b=float(input("Enter the left side of triangle :"))
c=float(input("Enter the right side of triangle :"))
print("The perimeter of triangle is :",triangle_peri(a,b,c))

B=float(input("Enter the one side of rectangle :"))
H=float(input("Enter the another side of rectangle :"))
print("The area of rectangle is :",rectangle_area(B,H))
print("The peri of rectangle is :",rectangle_peri(B,H))
```

```python
'''4. write a python program to find second largest number from
       the given sequenceof integer without using built-in function'''


def second_largest(list):
    maximum= max(list[0],list[1])
    second_max= min(list[0],list[1])
    n=len(list)
    for i in range(2, n):
        if list[i]>maximum:
            second_max=maximum
            maximum=list[i]
        else:
            if list[i]>second_max:
                second_max=list[i]

    return second_max

li=[]
n=int(input("Enter size of list "))
for i in range(0,n):
    e=int(input("Enter element of list "))
    li.append(e)

print("second largest in ",li,"is")
print(second_largest(li))
```

```python
'''5. write a python program to find the sum of the series
      1+(1/1!)+(1/2!)+(1/3!)+...+(1/n!)'''


def factorial(n):
    res = 1
    for i in range(2, n + 1):
            res *= i
    return res

def sum(n):
    s = 1

    for i in range(1, n + 1):
        s += 1.0 / factorial(i)
    print(s)

n = int(input("Enter the n value"))

sum(n)
```

```python
'''8.Write a python program to create, concatenate and print a string and perform the
    following operations.
        a. Access a substring
        b. Display the length of given string
        c. Demonstrate slicing and indexing in string
        d. Demonstrate string comparison'''


def con_str():
    str1 = input("Enter the first string: ")
    str2 = input("Enter the second string: ")
    result = str1 + str2
    print("Concatenated string:", result)

def acc_sbstr():
    original_str = input("Enter a string: ")
    start_index = int(input("Enter the start index: "))
    end_index = int(input("Enter the end index: "))
    substring = original_str[start_index:end_index]
    print("Substring:", substring)

def len_str():
    input_str = input("Enter a string: ")
    length = len(input_str)
    print("Length of the string:", length)

def slc_indx():
    input_str = input("Enter a string: ")
    index = int(input("Enter an index: "))
    sliced_char = input_str[:index]
    print("Character at index {}: {}".format(index, sliced_char))

def str_comp():
    str1 = input("Enter the first string: ")
    str2 = input("Enter the second string: ")
    if str1 == str2:
        print("Strings are equal.")
    else:
        print("Strings are not equal.")

while True:
    print("\nMenu:")
    print("1. Concatenate Strings")
    print("2. Access Substring")
    print("3. Length of String")
    print("4. Slicing and Indexing")
    print("5. String Comparison")
    print("6. Exit")

    choice = input("Enter your choice (1-6): ")

    if choice == '1':
        con_str()
    elif choice == '2':
        acc_sbstr()
    elif choice == '3':
        len_str()
    elif choice == '4':
        slc_indx()
    elif choice == '5':
        str_comp()
    elif choice == '6':
        print("Exiting the program. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 6.")
```

```python
class Student:
    def __init__(self, name, reg_no, course):
        self.name = name
        self.reg_no = reg_no
        self.course = course
        self.marks = []

    def add_marks(self, subject, mark):
        self.marks.append((subject, mark))

    def calculate_total(self):
        return sum(mark for _, mark in self.marks)

    def calculate_average(self):
        return self.calculate_total() / len(self.marks)

    def calculate_grade(self):
        average = self.calculate_average()
        if average >= 90:
            return 'A'
        elif 80 <= average < 90:
            return 'B'
        elif 70 <= average < 80:
            return 'C'
        elif 60 <= average < 70:
            return 'D'
        else:
            return 'F'

    def display_details(self):
        print(f"Name: {self.name}")
        print(f"Registration Number: {self.reg_no}")
        print(f"Course: {self.course}")
        print("Subject Marks:")
        for subject, mark in self.marks:
            print(f"  {subject}: {mark}")
        print(f"Total: {self.calculate_total()}")
        print(f"Average: {self.calculate_average()}")
        print(f"Grade: {self.calculate_grade()}")


def main():
    name = input("Enter student name: ")
    reg_no = input("Enter registration number: ")
    course = input("Enter course: ")

    student = Student(name, reg_no, course)

    for i in range(5):
        subject = input(f"Enter subject {i + 1} name: ")
        mark = float(input(f"Enter {subject} mark: "))
        student.add_marks(subject, mark)

    print("\nStudent Details:")
    student.display_details()


if __name__ == "__main__":
    main()
```

```python
class Employee:
    def __init__(self, emp_id, name, designation, basic_salary):
        self.emp_id = emp_id
        self.name = name
        self.designation = designation
        self.basic_salary = basic_salary

    def calculate_gross_salary(self):
        da = 0.05 * self.basic_salary
        hra = 0.10 * self.basic_salary
        gross_salary = self.basic_salary + da + hra
        return gross_salary

    def calculate_net_salary(self):
        da = 0.05 * self.basic_salary
        hra = 0.10 * self.basic_salary
        lic_deduction = 0.03 * self.basic_salary
        pf_deduction = 0.08 * self.basic_salary

        gross_salary = self.calculate_gross_salary()
        net_salary = gross_salary - (lic_deduction + pf_deduction)

        return net_salary

    def display_details(self):
        print(f"Employee ID: {self.emp_id}")
        print(f"Name: {self.name}")
        print(f"Designation: {self.designation}")
        print(f"Basic Salary: {self.basic_salary}")
        print(f"Gross Salary: {self.calculate_gross_salary()}")
        print(f"Net Salary: {self.calculate_net_salary()}")


def main():
    emp_id = input("Enter employee ID: ")
    name = input("Enter employee name: ")
    designation = input("Enter employee designation: ")
    basic_salary = float(input("Enter basic salary: "))

    employee = Employee(emp_id, name, designation, basic_salary)


    print("\nEmployee Details:")
    employee.display_details()


if __name__ == "__main__":
    main()
```
