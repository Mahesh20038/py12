```python
import math

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        area = math.pi * self.radius**2
        return area

radius = float(input("Enter the radius of the circle: "))

circle_instance = Circle(radius)

area = circle_instance.calculate_area()
print(f"The area of the circle with radius {radius} is: {area}")
```

```python
class Teacher:
    def __init__(self, name, designation, subject):
        self.name = name
        self.designation = designation
        self.subject = subject

    def display_information(self):
        print(f"Name: {self.name}")
        print(f"Designation: {self.designation}")
        print(f"Subject: {self.subject}")

class Professor(Teacher):
    def __init__(self, name, subject):
        super().__init__(name, "Professor", subject)

class AssistantProfessor(Teacher):
    def __init__(self, name, subject):
        super().__init__(name, "Assistant Professor", subject)

class AssociateProfessor(Teacher):
    def __init__(self, name, subject):
        super().__init__(name, "Associate Professor", subject)

# Example usage:
professor = Professor("John Doe", "Computer Science")
assistant_professor = AssistantProfessor("Jane Smith", "Physics")
associate_professor = AssociateProfessor("Alice Johnson", "Mathematics")

print("Professor Information:")
professor.display_information()

print("\nAssistant Professor Information:")
assistant_professor.display_information()

print("\nAssociate Professor Information:")
associate_professor.display_information()
```

```python
import os
import shutil

def get_current_directory():
    print(f"Current Working Directory: {os.getcwd()}")

def list_files_and_directories():
    files_and_directories = os.listdir()
    print("\nFiles and Directories in the Current Directory:")
    for item in files_and_directories:
        print(item)

def create_new_directory():
    new_directory_name = input("Enter the name of the new directory: ")
    os.mkdir(new_directory_name)
    print(f"Directory '{new_directory_name}' created successfully.")

def verify_directory_existence():
    directory_name = input("Enter the name of the directory to verify: ")
    if os.path.exists(directory_name) and os.path.isdir(directory_name):
        print(f"The directory '{directory_name}' exists.")
    else:
        print(f"The directory '{directory_name}' does not exist.")

def copy_file_to_directory():
    source_file = input("Enter the name of the file to copy: ")
    destination_directory = input("Enter the name of the destination directory: ")
    shutil.copy(source_file, destination_directory)
    print(f"File '{source_file}' copied to '{destination_directory}'.")

# Menu
while True:
    print("\nMenu:")
    print("1. Get Current Working Directory")
    print("2. List Files and Directories")
    print("3. Create a New Directory")
    print("4. Verify Directory Existence")
    print("5. Copy a File to Directory")
    print("6. Quit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        get_current_directory()
    elif choice == 2:
        list_files_and_directories()
    elif choice == 3:
        create_new_directory()
    elif choice == 4:
        verify_directory_existence()
    elif choice == 5:
        copy_file_to_directory()
    elif choice == 6:
        break
    else:
        print("Invalid choice. Please enter a valid option.")
```

```python
import os

def load_file():
    file_path = input("Enter the full path of the file to load: ")
    try:
        with open(file_path, 'r') as file:
            content = file.read()
            print(f"\nContents of the file:\n{content}")
    except FileNotFoundError:
        print(f"File not found at '{file_path}'.")

def display_file_contents():
    file_path = input("Enter the full path of the file to display: ")
    try:
        with open(file_path, 'r') as file:
            content = file.read()
            print(f"\nContents of the file:\n{content}")
    except FileNotFoundError:
        print(f"File not found at '{file_path}'.")

def rename_file():
    old_file_path = input("Enter the current full path of the file: ")
    new_file_name = input("Enter the new name for the file: ")

    try:
        os.rename(old_file_path, new_file_name)
        print(f"File renamed successfully to '{new_file_name}'.")
    except FileNotFoundError:
        print(f"File not found at '{old_file_path}'.")

def delete_file():
    file_path = input("Enter the full path of the file to delete: ")
    try:
        os.remove(file_path)
        print(f"File '{file_path}' deleted successfully.")
    except FileNotFoundError:
        print(f"File not found at '{file_path}'.")

# Menu
while True:
    print("\nMenu:")
    print("a. Load File")
    print("b. Display File Contents")
    print("c. Rename File")
    print("d. Delete File")
    print("q. Quit")

    choice = input("Enter your choice: ")

    if choice == 'a':
        load_file()
    elif choice == 'b':
        display_file_contents()
    elif choice == 'c':
        rename_file()
    elif choice == 'd':
        delete_file()
    elif choice == 'q':
        break
    else:
        print("Invalid choice. Please enter a valid option.")
```

```python
import re
from collections import Counter

def find_most_common_words(input_data, num_words):
    # If input_data is a file path, read the content from the file
    if os.path.isfile(input_data):
        with open(input_data, 'r', encoding='utf-8') as file:
            content = file.read()
    else:
        content = input_data

    # Tokenize the words using regular expression
    words = re.findall(r'\b\w+\b', content.lower())

    # Count the occurrences of each word
    word_counts = Counter(words)

    # Get the most common words and their frequencies
    most_common = word_counts.most_common(num_words)

    return most_common

# Example usage:
sample_text = """
This is a sample text. The purpose of this text is to provide an example for the find_most_common_words function.
It will count the occurrences of each word and return a list of tuples representing the most common words.
"""

result = find_most_common_words(sample_text, 5)
print("Most Common Words:")
for word, count in result:
    print(f"{word}: {count}")
```

```python
import os
import shutil

def display_menu():
    print("Menu:")
    print("1. List files in the external directory")
    print("2. Load a file from the external directory")
    print("3. Exit")

def list_files(external_directory):
    print("Files in the external directory:")
    files = os.listdir(external_directory)
    for file in files:
        print(file)

def load_file(external_directory):
    file_name = input("Enter the name of the file to load: ")
    source_path = os.path.join(external_directory, file_name)

    if os.path.exists(source_path):
        destination_path = os.path.join(os.getcwd(), file_name)
        shutil.copyfile(source_path, destination_path)
        print(f"File '{file_name}' loaded successfully.")
    else:
        print(f"File '{file_name}' not found in the external directory.")

if __name__ == "__main__":
    external_directory = "D:\praveen"

    while True:
        display_menu()
        choice = input("Enter your choice (1-3): ")

        if choice == "1":
            list_files(external_directory)
        elif choice == "2":
            load_file(external_directory)
        elif choice == "3":
            print("Exiting the program. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter a valid option.")
```
