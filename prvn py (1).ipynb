```python
def copy_alternative_lines(file1, file2):
    try:
        with open(file1, 'r') as f1, open(file2, 'w') as f2:
            lines = f1.readlines()

            # Using list slicing to get odd-numbered lines
            odd_lines = lines[::2]

            # Writing odd-numbered lines to file2
            f2.writelines(odd_lines)

        print(f"Odd-numbered lines copied from {file1} to {file2} successfully.")

    except FileNotFoundError:
        print(f"Error: File '{file1}' not found.")

    except PermissionError:
        print(f"Error: Permission issue while accessing {file1} or {file2}.")

    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage
file1 ="C:\\Users\\User\\Desktop\\txt1.txt"
file2 = "C:\\Users\\User\\Desktop\\txt2.txt"
copy_alternative_lines(file1, file2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_curves():
    # Generate x values
    x = np.linspace(-2*np.pi, 2*np.pi, 1000)

    # Sine curve
    y_sine = np.sin(x)

    # Cosine curve
    y_cosine = np.cos(x)

    # Polynomial curve (example: y = x^3 - 6x^2 + 11x - 6)
    y_polynomial = x**3 - 6*x**2 + 11*x - 6

    # Exponential curve (example: y = e^x)
    y_exponential = np.exp(x)

    # Plotting the curves
    plt.figure(figsize=(10, 6))

    plt.subplot(2, 2, 1)
    plt.plot(x, y_sine, label='Sine')
    plt.title('Sine Curve')
    plt.legend()

    plt.subplot(2, 2, 2)
    plt.plot(x, y_cosine, label='Cosine', color='orange')
    plt.title('Cosine Curve')
    plt.legend()

    plt.subplot(2, 2, 3)
    plt.plot(x, y_polynomial, label='Polynomial', color='green')
    plt.title('Polynomial Curve')
    plt.legend()

    plt.subplot(2, 2, 4)
    plt.plot(x, y_exponential, label='Exponential', color='red')
    plt.title('Exponential Curve')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Call the function to display the curves
plot_curves()
```

```python
'''Consider a showroom of electronic products, where there are various salesmen. Each salesman is given a commission of 5%,
depending on the sales made per month. In case the sale done is less than 50000, then the salesman is not given any commission.
Write a function to calculate total sales of a salesman in a month, commission and remarks for the salesman. Sales done by
each salesman per week is to be provided as input. Use tuples / list to store data of salesmen. Assign remarks according
to the following criteria:
Excellent: Sales >=80000
Good: Sales>=60000 and <80000
Average: Sales>=40000 and <60000
Work Hard: Sales < 40000'''



def calculate_sales_commission(sales_per_week):
    total_sales = sum(sales_per_week)
    commission_rate = 0.05  # 5%

    if total_sales >= 50000:
        commission = total_sales * commission_rate

        if total_sales >= 80000:
            remarks = "Excellent"
        elif 60000 <= total_sales < 80000:
            remarks = "Good"
        elif 40000 <= total_sales < 60000:
            remarks = "Average"
        else:
            remarks = "Work Hard"
    else:
        commission = 0
        remarks = "No commission for sales less than 50000"

    return total_sales, commission, remarks

def main():
    # Example data: Sales per week for each salesman
    salesmen_data = [
        ("John", [12000, 15000, 20000, 18000]),
        ("Alice", [25000, 30000, 22000, 18000]),
        ("Bob", [35000, 40000, 38000, 42000]),
    ]

    for salesman, sales_per_week in salesmen_data:
        total_sales, commission, remarks = calculate_sales_commission(sales_per_week)

        print(f"Salesman: {salesman}")
        print(f"Total Sales: {total_sales}")
        print(f"Commission: {commission}")
        print(f"Remarks: {remarks}")
        print("\n")

if __name__ == "__main__":
    main()
```

```python
def calculate_sales_commission(sales_per_week):
    total_sales = sum(sales_per_week)
    commission_rate = 0.05  # 5% commission rate

    if total_sales >= 50000:
        commission = total_sales * commission_rate
        if total_sales >= 80000:
            remarks = "Excellent"
        elif total_sales >= 60000:
            remarks = "Good"
        elif total_sales >= 40000:
            remarks = "Average"
        else:
            remarks = "Work Hard"
    else:
        commission = 0
        remarks = "No Commission (Sales < 50000)"

    return total_sales, commission, remarks

def main():
    num_salesmen = int(input("Enter the number of salesmen: "))
    salesmen_data = []

    for i in range(1, num_salesmen + 1):
        salesman_name = input(f"Enter the name of salesman {i}: ")
        sales_per_week = [int(input(f"Enter sales for week {j + 1}: ")) for j in range(4)]  # Assuming 4 weeks in a month
        salesman_info = (salesman_name, *calculate_sales_commission(sales_per_week))
        salesmen_data.append(salesman_info)

    # Displaying results
    for name, total_sales, commission, remarks in salesmen_data:
        print(f"\nSalesman: {name}")
        print(f"Total Sales: ${total_sales}")
        print(f"Commission: ${commission}")
        print(f"Remarks: {remarks}")

if __name__ == "__main__":
    main()

```

```python
#pip install matplotlib
import matplotlib.pyplot as plt

def display_histogram(numbers):
    plt.hist(numbers, bins=10, edgecolor='black', alpha=0.7)
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.title('Histogram')
    plt.show()

def main():
    try:
        # Input the number of integers
        n = int(input("Enter the number of integers: "))

        # Input the list of integers
        numbers = []
        for i in range(n):
            num = int(input(f"Enter integer {i + 1}: "))
            numbers.append(num)

        # Display the histogram
        display_histogram(numbers)

    except ValueError:
        print("Error: Please enter valid integers.")

if __name__ == "__main__":
    main()
```
